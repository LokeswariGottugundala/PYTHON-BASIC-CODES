# PRIME NUMBER OR NOT

```python
num = int(input())
flag = 0
for i in range(2,num):
  if num%i==0:
    flag = 1
    break
if flag == 1:
  print('Not Prime')
else:
  print("Prime")
```

# SUM OF ALL DIGITS IN A NUM

```python
n = int(input())
sum=0
while n >0:
  sum+=int(n%10)
  n=n//10
print(sum)
```

# PALINDROME

```python
n=int(input())
rev=int(str(n)[::-1])
if n==rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

# ARMSTRONG  NUM

```python
n=int(input())
sum = 0
temp = n
while temp > 0:
  rem = temp%10
  sum += rem**3
  temp //=10
if n == sum:
  print("armstrong")
else:
  print("not armstrong")
```

# ARMSTRONG NUM WITHIN A RANGE

```python
def armstrong(n1,n2):

  for n in range(n1,n2+1):
    sum = 0
    temp = n
    while temp > 0:
      rem = temp%10
      sum += rem**3
      temp //=10

    if n ==sum:
      print(n,end=" ")
n1=int(input())
n2= int(input())
armstrong(n1,n2)
```

#FIBANOCCI SERIES

```python
n = int(input())
a = 0
b = 1
print(a,b,end=" ")
for i in range(2,n):
  c = a + b
  a = b
  b = c
  print(c,end=" ")
```

# STRONG NUMBER

```python
n = int(input())
temp = n
sum = 0
while n>0:
  digit = n%10
  n =n//10
  fact =1
  for i in range(1,digit+1):
    fact = fact*i
  sum +=fact
if sum == temp:
  print("strong number")
else:
  print("not a strong number")
```
