# SUM OF NUMBERS IN GIVEN RANGE

```python
n1=int(input())
n2=int(input())
sum=0
for i in range(n1,n2+1):
  sum +=i
print(sum)
```

# PRIME NUMBER OR NOT

```python
num = int(input())
flag = 0
for i in range(2,num):
  if num%i==0:
    flag = 1
    break
if flag == 1:
  print('Not Prime')
else:
  print("Prime")
```

```python
n = int(input())
for i in range(2,n):
  if n%i==0:
    print("Not Prime")
    break
else:
  print("Prime")
```

# PRIME NUM IN GIVEN RANGE

```python
n1=int(input())
n2=int(input())
for i in range(n1,n2+1):
  for j in range(2,i):
    if i%j==0:
      break
  else:
    print(i,end=" ")
```

# SUM OF ALL DIGITS IN A NUM

```python
n = int(input())
sum=0
while n >0:
  sum+=int(n%10)
  n=n//10
print(sum)
```

# REVERSE NUM

```python
n=int(input())
print(str(n)[::-1])
```

```python
n= int(input())
rev=0
while n>0:
  rem = n %10
  rev=rev*10+rem             #same logic for palindrome
  n = n//10
print(rev)
```

# PALINDROME

```python
n=int(input())
rev=int(str(n)[::-1])
if n==rev:
  print("Palindrome")
else:
  print("Not Palindrome")
```

# ARMSTRONG  NUM

```python
n=int(input())
sum = 0
temp = n
while temp > 0:
  rem = temp%10
  sum += rem**3
  temp //=10
if n == sum:
  print("armstrong")
else:
  print("not armstrong")
```

# ARMSTRONG NUM WITHIN A RANGE

```python
def armstrong(n1,n2):

  for n in range(n1,n2+1):
    sum = 0
    temp = n
    while temp > 0:
      rem = temp%10
      sum += rem**3
      temp //=10

    if n ==sum:
      print(n,end=" ")
n1=int(input())
n2= int(input())
armstrong(n1,n2)
```

FIBANOCCI SERIES

```python
n = int(input())
a = 0
b = 1
print(a,b,end=" ")
for i in range(2,n):
  c = a + b
  a = b
  b = c
  print(c,end=" ")
```
