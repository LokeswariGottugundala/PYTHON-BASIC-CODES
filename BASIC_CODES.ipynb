# PRIME NUMBER OR NOT

```python
num = int(input())
flag = 0
for i in range(2,num):
  if num%i==0:
    flag = 1
    break
if flag == 1:
  print('Not Prime')
else:
  print("Prime")
```

# ARMSTRONG  NUM

```python
n=int(input())
sum = 0
temp = n
while temp > 0:
  rem = temp%10
  sum += rem**3
  temp //=10
if n == sum:
  print("armstrong")
else:
  print("not armstrong")
```

# ARMSTRONG NUM WITHIN A RANGE

```python
def armstrong(n1,n2):

  for n in range(n1,n2+1):
    sum = 0
    temp = n
    while temp > 0:
      rem = temp%10
      sum += rem**3
      temp //=10

    if n ==sum:
      print(n,end=" ")
n1=int(input())
n2= int(input())
armstrong(n1,n2)
```

# STRONG NUMBER

```python
n = int(input())
temp = n
sum = 0
while n>0:
  digit = n%10
  n =n//10
  fact =1
  for i in range(1,digit+1):
    fact = fact*i
  sum +=fact
if sum == temp:
  print("strong number")
else:
  print("not a strong number")
```

# PERFECT NUMBER

#FRIENDLY PAIR/AMICABLE PAIR

```python
num1 = int(input())
num2 = int(input())

sum1 = 0
sum2 = 0
for i in range(1, num1):
    if num1 % i == 0:
        sum1 += i
for i in range(1, num2):
    if num2 % i == 0:
        sum2 += i
if sum1 == num1 and sum2 == num2:
    print("friendly pair")
else:
    print("not a friendly pair")
```

#HCF OF A NUMBER

```python
n1=int(input())
n2=int(input())
hcf=1
for i  in range(1,min(n1,n2)):
  if n1%i==0 and n2%i==0:
    hcf=i
print(hcf)
```
