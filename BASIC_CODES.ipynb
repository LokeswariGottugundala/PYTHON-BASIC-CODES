# ARMSTRONG  NUM

```python
n=int(input())
sum = 0
temp = n
while temp > 0:
  rem = temp%10
  sum += rem**3
  temp //=10
if n == sum:
  print("armstrong")
else:
  print("not armstrong")
```

# ARMSTRONG NUM WITHIN A RANGE

```python
def armstrong(n1,n2):

  for n in range(n1,n2+1):
    sum = 0
    temp = n
    while temp > 0:
      rem = temp%10
      sum += rem**3
      temp //=10

    if n ==sum:
      print(n,end=" ")
n1=int(input())
n2= int(input())
armstrong(n1,n2)
```

#FRIENDLY PAIR/AMICABLE PAIR

```python
num1 = int(input())
num2 = int(input())

sum1 = 0
sum2 = 0
for i in range(1, num1):
    if num1 % i == 0:
        sum1 += i
for i in range(1, num2):
    if num2 % i == 0:
        sum2 += i
if sum1 == num1 and sum2 == num2:
    print("friendly pair")
else:
    print("not a friendly pair")
```

#2ND SMALLEST ELEMENT

```python

def find_2nd_smallest(arr):
    if len(arr) < 2:
        return "Array must have at least two elements"

    unique_arr = sorted(list(set(arr)))

    if len(unique_arr) < 2:
        return "No second smallest element (all elements are the same)"

    return unique_arr[1]
```

#FREQUENCY OF ELEMENTS

```python
def count_frequency(arr):
    counted_elements = []
    for i in range(len(arr)):
        if arr[i] not in counted_elements:
            count = 1
            for j in range(i + 1, len(arr)):
                if arr[i] == arr[j]:
                    count += 1
            counted_elements.append(arr[i])
            print(f"{arr[i]}: {count}")
arr =input().split()
count_frequency(arr)
```

```python
def frequency_of_elements(arr):
    frequency = {}
    for element in arr:
        if element in frequency:
            frequency[element] += 1
        else:
            frequency[element] = 1
    return frequency
arr=input().split()
print(frequency_of_elements(arr))
```
